## Main Goal:  
We have a target binary outcome $A$, we are given user characteristics $\vec x$ and possible interventions $\vec y$.  
We build a model to find optimal interventions to complement the decision process by:

$$\boxed{
  \begin{aligned}
  &\forall \vec x \text{ and } \vec x' = \vec x - \vec y: \\
  &\text{arg max}_{\vec y}\left\{[P(A|\vec x, \space \text{do}(\vec y) ) - P(A|\vec x)]\cdot [\text{cost}(\vec x, A = \text{True}) - \text{cost}(\vec x, A = \text{False})] \right\}
  \end{aligned}
}$$

for however many $\vec y$ we'd like depending on threshold and buisness capacity.
